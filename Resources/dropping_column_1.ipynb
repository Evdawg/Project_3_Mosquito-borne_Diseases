```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
precip_data_to_convert = Path("Data/Clean_county_avg_precipitation.csv")
temp_data_to_convert = Path("Data/Clean_county_avg_temperature.csv")
LD_data_to_convert = Path("Data/Clean_Ld-Case-Counts-by-County-01-20.csv")
WN_data_to_convert = Path("Data/Clean_WestNile-Case-Counts-by-County-01-20.csv")

# Read Data File and store into Pandas DataFrames
precip_data_ = pd.read_csv(precip_data_to_convert)
temp_data_ = pd.read_csv(temp_data_to_convert)
LD_data_ = pd.read_csv(LD_data_to_convert)
WN_data_ = pd.read_csv(WN_data_to_convert)

precip_data_.head()
```

```python
# Drop the first column from precip_data_
precip_data_.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Export the Precip DataFrame to a new Precip CSV file
output_csv_file = "Data/precip_data.csv"
precip_data_.to_csv(output_csv_file, index=False)

print(f"The Precipitation DataFrame has been exported to {output_csv_file}.")
```

```python
# Export the Temp DataFrame to a new Temp CSV file
output_csv_file = "Data/temp_data.csv"
temp_data_.to_csv(output_csv_file, index=False)

print(f"The Temperature DataFrame has been exported to {output_csv_file}.")
```

```python
# Export the LD DataFrame to a new LD CSV file
output_csv_file = "Data/LD_data.csv"
LD_data_.to_csv(output_csv_file, index=False)

print(f"The LD DataFrame has been exported to {output_csv_file}.")
```

```python
# Export the LD DataFrame to a new LD CSV file
output_csv_file = "Data/WN_data.csv"
WN_data_.to_csv(output_csv_file, index=False)

print(f"The WN DataFrame has been exported to {output_csv_file}.")
```

```python
# Extract 'state' and 'county' from the split column and create separate columns for them
WN_data_['county'] = WN_data_['county'].str.split(', ')
WN_data_[['state', 'county']] = pd.DataFrame(WN_data_['county'].to_list(), index=WN_data_.index)

# Display the updated DataFrame
WN_data_.head()
```

```python
# Replace all NaN values with 0
WN_data_.fillna(0, inplace=True)

# Display the updated DataFrame
WN_data_.head()
```

```python
#move state column
WN_data_1 = WN_data_[['county', 'state'] + list(WN_data_.columns[1:-2]) + ['lat', 'lon', 'FIPS']]

# Display the updated DataFrame
WN_data_1.head()
```

```python
output_csv_file = "Data/WN_data_vis.csv"
WN_data_1.to_csv(output_csv_file, index=False)
```
